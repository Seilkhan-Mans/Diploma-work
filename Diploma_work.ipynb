# Random Over Sampling

As I have already mentioned, classes are highly imbalanced. One of the ways to eliminate the imbalance is OverSampling. This method fills a small class with an artificial one.

```python
from imblearn.over_sampling import  RandomOverSampler, SMOTE
oversample=RandomOverSampler(sampling_strategy=1)
```

```python
oversample
```

```python
x_over, y_over=oversample.fit_resample(x,y)
```

```python
x_over.shape, y_over.shape
```

```python
y_over.value_counts().head()
ax=sns.countplot(y_over)
Target_temp = y_over.value_counts()
print(Target_temp)
plt.tight_layout()
plt.show()
```

```python
def smooth_curve(points, factor=0.9):
    smoothed_points = []
    for point in points:
        if smoothed_points:
            previous = smoothed_points[-1]
            smoothed_points.append(previous * factor + point * (1 - factor))
        else:
                smoothed_points.append(point)
    return smoothed_points 
smooth_accuracy_history_test = smooth_curve(ave_mse_history_test[10:])
smooth_accuracy_history_valid = smooth_curve(ave_mse_history_valid[10:])
plt.plot(range(1, len(smooth_accuracy_history_test) + 1), smooth_accuracy_history_test)
plt.plot(range(1, len(smooth_accuracy_history_valid) + 1), smooth_accuracy_history_valid)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
def smooth_curve(points, factor=0.9):
    smoothed_points = []
    for point in points:
        if smoothed_points:
            previous = smoothed_points[-1]
            smoothed_points.append(previous * factor + point * (1 - factor))
        else:
                smoothed_points.append(point)
    return smoothed_points 
smooth_mse_history_test = smooth_curve(ave_mse_history_test[10:])
smooth_mse_history_valid = smooth_curve(ave_mse_history_valid[10:])
plt.plot(range(1, len(smooth_mse_history_test) + 1), smooth_mse_history_test)
plt.plot(range(1, len(smooth_mse_history_valid) + 1), smooth_mse_history_valid)
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
